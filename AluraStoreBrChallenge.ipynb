#1. Análise do faturamento


```python
import pandas as pd


url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"


loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)


def analisar_loja(df, nome):

    df.columns = df.columns.str.strip()

    tamanho = len(df)
    faturamento = df['Preço'].sum()
    maiores_vendas = df['Preço'].nlargest(5).tolist()
    total_maiores_vendas = sum(maiores_vendas)

    print(f"Loja {nome}:")
    print(f"- Total de vendas: {tamanho}")
    print(f"- Faturamento total: R$ {faturamento:.2f}")
    print(f"- 5 maiores vendas: {maiores_vendas}\n")
    print(f"Total das 5 maiores vendas: R$ {total_maiores_vendas:.2f}")


analisar_loja(loja1, "1")
analisar_loja(loja2, "2")
analisar_loja(loja3, "3")
analisar_loja(loja4, "4")

```

```python
import pandas as pd
import matplotlib.pyplot as plt


url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"


lojas = [
    pd.read_csv(url1),
    pd.read_csv(url2),
    pd.read_csv(url3),
    pd.read_csv(url4)
]

for df in lojas:
    df.columns = df.columns.str.strip()

faturamentos = [df['Preço'].sum() for df in lojas]
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']


plt.figure(figsize=(8, 5))
plt.bar(nomes_lojas, faturamentos, color='skyblue')
plt.title('Faturamento Total por Loja')
plt.ylabel('Faturamento (R$)')
plt.xlabel('Lojas')


for i, valor in enumerate(faturamentos):
    plt.text(i, valor + 50, f'R$ {valor:,.2f}', ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
import seaborn as sns


urls = [url1, url2, url3, url4]
lojas = [pd.read_csv(url) for url in urls]

for df in lojas:
    df.columns = df.columns.str.strip()


for i, loja in enumerate(lojas, start=1):
    categoria_faturamento = loja.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)
    print(f"Loja {i} - Faturamento por categoria:\n{categoria_faturamento}\n")


    plt.figure(figsize=(8, 4))
    sns.barplot(x=categoria_faturamento.values, y=categoria_faturamento.index, palette='viridis')
    plt.title(f'Loja {i} - Faturamento por Categoria')
    plt.xlabel('Faturamento (R$)')
    plt.ylabel('Categoria')
    plt.tight_layout()
    plt.show()


todas_lojas = pd.concat(lojas)
todas_lojas.columns = todas_lojas.columns.str.strip()
categoria_total = todas_lojas.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)

print("Faturamento total por categoria (todas as lojas):")
print(categoria_total)


plt.figure(figsize=(8, 5))
sns.barplot(x=categoria_total.values, y=categoria_total.index, palette='magma')
plt.title('Faturamento Total por Categoria (Todas as Lojas)')
plt.xlabel('Faturamento (R$)')
plt.ylabel('Categoria')
plt.tight_layout()
plt.show()
```
